# Matrix Operations & Linear Algebra Toolkit

```python
import numpy as np

def add_matrices(A, B):
    return A + B

def subtract_matrices(A, B):
    return A - B

def multiply_matrices(A, B):
    return A @ B

def determinant(A):
    return np.linalg.det(A)

def inverse(A):
    try:
        return np.linalg.inv(A)
    except np.linalg.LinAlgError:
        return "Matrix is singular, cannot invert."

def eigenvalues(A):
    return np.linalg.eigvals(A)

def eigenvectors(A):
    return np.linalg.eig(A)[1]


# -------- Main Program --------

A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

print("Matrix A:\n", A)
print("Matrix B:\n", B)

print("\nAddition:\n", add_matrices(A, B))
print("\nSubtraction:\n", subtract_matrices(A, B))
print("\nMultiplication:\n", multiply_matrices(A, B))

print("\nDeterminant of A:\n", determinant(A))

print("\nInverse of A:\n", inverse(A))

print("\nEigenvalues of A:\n", eigenvalues(A))

print("\nEigenvectors of A:\n", eigenvectors(A))
```

# 1.Jacobi Method 
# 2.Gauss-Seidel Method
# 3.Gauss Elimination Method

```python
import numpy as np
import matplotlib.pyplot as plt

# --------------------------
# User Input or Example
# --------------------------
def get_system():
    use_input = input("Do you want to enter your own matrix? (y/n): ").lower()
    if use_input == 'y':
        n = int(input("Enter size of system (n): "))
        A = np.zeros((n,n))
        b = np.zeros(n)
        print("Enter matrix A row by row:")
        for i in range(n):
            for j in range(n):
                A[i][j] = float(input(f"A[{i},{j}] = "))
            b[i] = float(input(f"b[{i}] = "))
        return A, b
    else:
        # Example system
        A = np.array([[4,1,2],
                      [3,5,1],
                      [1,1,3]], dtype=float)
        b = np.array([4,7,3], dtype=float)
        return A, b

# --------------------------
# Jacobi Method
# --------------------------
def jacobi(A, b, x0, tol=1e-6, max_iter=100):
    n = len(b)
    x = x0.copy()
    errors = []

    for k in range(max_iter):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j]*x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s)/A[i][i]

        error = np.linalg.norm(x_new - x)
        errors.append(error)

        if error < tol:
            break

        x = x_new

    return x, errors

# --------------------------
# Gauss-Seidel Method
# --------------------------
def gauss_seidel(A, b, x0, tol=1e-6, max_iter=100):
    n = len(b)
    x = x0.copy()
    errors = []

    for k in range(max_iter):
        x_old = x.copy()
        for i in range(n):
            s1 = sum(A[i][j]*x[j] for j in range(i))
            s2 = sum(A[i][j]*x_old[j] for j in range(i+1, n))
            x[i] = (b[i] - s1 - s2)/A[i][i]

        error = np.linalg.norm(x - x_old)
        errors.append(error)

        if error < tol:
            break

    return x, errors

# --------------------------
# Gauss Elimination Method
# --------------------------
def gauss_elimination(A, b):
    A = A.astype(float)
    b = b.astype(float)
    n = len(b)

    # Forward elimination
    for k in range(n-1):
        for i in range(k+1, n):
            factor = A[i][k]/A[k][k]
            for j in range(k, n):
                A[i][j] -= factor*A[k][j]
            b[i] -= factor*b[k]

    # Back substitution
    x = np.zeros(n)
    for i in range(n-1, -1, -1):
        s = sum(A[i][j]*x[j] for j in range(i+1, n))
        x[i] = (b[i] - s)/A[i][i]

    return x

# --------------------------
# Main Program
# --------------------------
A, b = get_system()
x0 = np.zeros(len(b))

# Solve using Jacobi
x_jacobi, err_jacobi = jacobi(A, b, x0)
print("\nJacobi Solution:", x_jacobi)

# Solve using Gauss-Seidel
x_gs, err_gs = gauss_seidel(A, b, x0)
print("Gauss-Seidel Solution:", x_gs)

# Solve using Gauss Elimination
x_ge = gauss_elimination(A.copy(), b.copy())
print("Gauss Elimination Solution:", x_ge)

# --------------------------
# Convergence Plot
# --------------------------
plt.plot(err_jacobi, marker='o', label='Jacobi')
plt.plot(err_gs, marker='x', label='Gauss-Seidel')
plt.yscale('log')
plt.xlabel('Iteration')
plt.ylabel('Error (norm)')
plt.title('Convergence Comparison')
plt.grid(True)
plt.legend()
plt.show()
```

## LU Decomposition
## QR Decomposition (Gram-Schmidt)
## Singular Value Decomposition (SVD)

```python
import numpy as np

# --------------------------
# Example matrix (default)
# --------------------------
A = np.array([[4,3,2],
              [2,1,3],
              [6,5,4]], dtype=float)

# Optional: Uncomment to take input from user
# n = int(input("Enter size of square matrix: "))
# A = np.zeros((n,n))
# for i in range(n):
#     for j in range(n):
#         A[i,j] = float(input(f"A[{i},{j}] = "))

print("Matrix A:\n", A)

# --------------------------
# LU Decomposition
# --------------------------
def lu_decomposition(A):
    A = A.astype(float)
    n = A.shape[0]
    L = np.zeros((n,n))
    U = np.zeros((n,n))

    for i in range(n):
        L[i,i] = 1
        for j in range(i, n):
            s = sum(L[i,k]*U[k,j] for k in range(i))
            U[i,j] = A[i,j] - s
        for j in range(i+1, n):
            s = sum(L[j,k]*U[k,i] for k in range(i))
            L[j,i] = (A[j,i] - s)/U[i,i]
    return L, U

L, U = lu_decomposition(A)
print("\nLU Decomposition:")
print("L:\n", L)
print("U:\n", U)
print("Check L @ U:\n", L @ U)

# --------------------------
# QR Decomposition (Gram-Schmidt)
# --------------------------
def qr_decomposition(A):
    A = A.astype(float)
    n, m = A.shape
    Q = np.zeros((n,m))
    R = np.zeros((m,m))

    for j in range(m):
        v = A[:,j]
        for i in range(j):
            R[i,j] = np.dot(Q[:,i], A[:,j])
            v = v - R[i,j]*Q[:,i]
        R[j,j] = np.linalg.norm(v)
        Q[:,j] = v / R[j,j]
    return Q, R

Q, R = qr_decomposition(A)
print("\nQR Decomposition:")
print("Q:\n", Q)
print("R:\n", R)
print("Check Q @ R:\n", Q @ R)

# --------------------------
# Singular Value Decomposition (SVD)
# --------------------------
U, S, Vt = np.linalg.svd(A)
print("\nSingular Value Decomposition (SVD):")
print("U:\n", U)
print("Singular values S:\n", S)
print("Vt:\n", Vt)

# Reconstruct A using SVD
Sigma = np.zeros_like(A, dtype=float)
for i in range(len(S)):
    Sigma[i,i] = S[i]

A_reconstructed = U @ Sigma @ Vt
print("Reconstructed A (U @ Sigma @ Vt):\n", A_reconstructed)
```
